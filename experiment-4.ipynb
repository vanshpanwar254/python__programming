###Q1Find a factorial of given number. 

```python
n = 5

factorial = 1

for i in range(1, n + 1):
    factorial = factorial * i

factorial
```

###Q2Find whether the given number is Armstrong number. 

```python

num = 153

temp = num
sum = 0

while temp > 0:
    digit = temp % 10
    sum = sum + digit**3
    temp = temp // 10

if sum == num:
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

###Q3Print Fibonacci series up to given term. 

```python
n = 10   # number of terms

a = 0
b = 1

print(a)
print(b)

for i in range(2, n):
    c = a + b
    print(c)
    a = b
    b = c
```

###Q4Write a program to find if given number is prime number or not. 

```python
num = 7

if num > 1:
    for i in range(2, num):
        if num % i == 0:
            print("Not a Prime Number")
            break
    else:
        print("Prime Number")
else:
    print("Not a Prime Number")
```

###Q5Check whether given number is palindrome or not. 

```python
num = 121

temp = num
reverse = 0

while temp > 0:
    digit = temp % 10
    reverse = reverse * 10 + digit
    temp = temp // 10

if reverse == num:
    print("Palindrome Number")
else:
    print("Not a Palindrome Number")
```

###Q6Write a program to print sum of digits. 

```python
num = 1234

temp = num
sum_of_digits = 0

while temp > 0:
    digit = temp % 10
    sum_of_digits = sum_of_digits + digit
    temp = temp // 10

sum_of_digits
```

###Q7Count and print all numbers divisible by 5 or 7 between 1 to 100. 

```python
count = 0

for num in range(1, 101):
    if num % 5 == 0 or num % 7 == 0:
        print(num)
        count = count + 1

print("Total numbers:", count)
```

###Q9Print all prime numbers between 1 and 100. 

```python
for num in range(2, 101):
    is_prime = True
    
    for i in range(2, num):
        if num % i == 0:
            is_prime = False
            break
    
    if is_prime:
        print(num)
```
