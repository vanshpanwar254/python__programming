###1.Add few names, one name in each row, in “name.txt file”. 
a.	Count no of names 
b.	Count all names starting with vowel 
c.	Find longest name 

```python
f = open("name.txt", "w")
f.write("Aman\n")
f.write("Riya\n")
f.write("Ishita\n")
f.write("Om\n")
f.write("Anjali\n")
f.write("Uday\n")
f.close()

f = open("name.txt", "r")
names = f.readlines()
f.close()

for i in range(len(names)):
    names[i] = names[i].strip()

count = len(names)
print("Total number of names:", count)

vowel_count = 0

for name in names:
    first_letter = name[0].upper()
    
    if first_letter == 'A' or first_letter == 'E' or first_letter == 'I' or first_letter == 'O' or first_letter == 'U':
        vowel_count = vowel_count + 1

print("Names starting with vowel:", vowel_count)

longest = names[0]

for name in names:
    if len(name) > len(longest):
        longest = name

print("Longest name:", longest)
```

###2.Store integers in a file. 
a.	Find the max number 
b.	Find average of all numbers 
c.	Count number of numbers greater than 100 


```python
f = open("numbers.txt", "w")
f.write("10\n")
f.write("150\n")
f.write("75\n")
f.write("200\n")
f.write("50\n")
f.write("120\n")
f.close()

f = open("numbers.txt", "r")
nums = f.readlines()
f.close()

for i in range(len(nums)):
    nums[i] = int(nums[i].strip())


max_num = nums[0]

for num in nums:
    if num > max_num:
        max_num = num

print("Maximum number:", max_num)

total = 0

for num in nums:
    total = total + num

average = total / len(nums)

print("Average of numbers:", average)

count = 0

for num in nums:
    if num > 100:
        count = count + 1

print("Numbers greater than 100:", count)
```

###3.Assume a file city.txt with details of 5 cities in given format (cityname population(in lakhs) area(in sq KM) ): 
Example: 
Dehradun 5.78 308.20 
Delhi 190 1484 
…………… 
Open file city.txt and read to: 
a.	Display details of all cities 
b.	Display city names with population more than 10Lakhs 
c.	Display sum of areas of all cities 


```python
f = open("city.txt", "w")
f.write("Dehradun 5.78 308.20\n")
f.write("Delhi 190 1484\n")
f.write("Mumbai 124 603\n")
f.write("Chandigarh 11 114\n")
f.write("Jaipur 30 467\n")
f.close()

f = open("city.txt", "r")
lines = f.readlines()
f.close()

print("Details of all cities:")
for line in lines:
    print(line.strip())

print("\nCities with population more than 10 Lakhs:")

for line in lines:
    data = line.split()
    city = data[0]
    population = float(data[1])
    
    if population > 10:
        print(city)


total_area = 0

for line in lines:
    data = line.split()
    area = float(data[2])
    total_area = total_area + area

print("\nTotal area of all cities:", total_area)
```

###4.Input two values from user where the first line contains N, the number of test cases. The next N lines contain the space separated values of a and b. Perform integer division and print a/b. Handle exception in case of ZeroDivisionError or ValueError.  
Sample input 
1 0 
2 $ 
3 1  
Sample Output: 
Error Code: integer division or modulo by zero  
Error Code: invalid literal for int() with base 10: '$' 3 


```python
N = 3
test_cases = ["1 0", "2 $", "3 1"]

for i in range(N):
    try:
        a, b = test_cases[i].split()

        a = int(a)
        b = int(b)
        
        result = a // b
        
        print(result)
    
    except ZeroDivisionError as e:
        print("Error Code:", e)
    
    except ValueError as e:
        print("Error Code:", e)
```

###5.Create multiple suitable exceptions for a file handling program.

```python
try:
    f = open("data.txt", "r")
    
    content = f.read()
    print("File Content:")
    print(content)
    
    f.close()

except FileNotFoundError:
    print("Error: File not found.")

except PermissionError:
    print("Error: Permission denied to open the file.")

except IOError:
    print("Error: Input/Output error occurred.")

except ValueError:
    print("Error: Invalid value found.")

except Exception as e:
    print("Some other error occurred:", e)

finally:
    print("Program Ended.")
```
