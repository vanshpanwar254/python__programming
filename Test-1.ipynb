##Q1

```python
def top_k(item, k):
    item.sort(key=lambda x: x[1], reverse=True)
    return item[:k]

item = [("A", 90), ("B", 95), ("C", 95), ("D", 85)]
k = 2

result = top_k(item, k)
print(result)
```

##Q2

```python
pairs = [
    ("fruit", "apple"),
    ("fruit", "banana"),
    ("fruit", "apple"),
    ("veg", "carrot"),
    ("veg", "carrot")
]

def count_by_category(pairs):
    d = {}
    for cat, word in pairs:
        if cat  not in d:
            d[cat]={}
            if word not in d[cat]:
                d[cat][word]= 0
                d[cat][word] +=1
                return d



print(count_by_category(pairs))
            
```

```python
def missing_number(nums):
    n = len(nums)
    total =n*(n+1) //2
    return total - sum(nums)

print(missing_number([3, 0, 1]))
```

##Q4

```python
def add_unique(x, seen=None):
    if seen is None:
        seen = set()
    
    if x not in seen:
        seen.add(x)
    
    return list(seen)

print(add_unique(5))       
print(add_unique(5, {5, 6}))     
print(add_unique(7, {5, 6}))     
         
```

```python
data = [1, 2, 3, 4, 5]
print(moving_average(data, 3))

def moving_average(stream, w):
    result = []
    current_sum = 0

    for i in range(len(stream)):
        current_sum += stream[i]

        # Remove element that goes out of window
        if i >= w:
            current_sum -= stream[i - w]

        # Calculate window size (for initial elements)
        window_size = min(i + 1, w)

        result.append(current_sum / window_size)

    return result
```
