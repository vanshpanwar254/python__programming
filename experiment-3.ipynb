###1.Check whether a given number is divisible by 3 and 5 both. 

```python
num = 15

if num % 3 == 0 and num % 5 == 0:
    print("Number is divisible by both 3 and 5")
else:
    print("Number is not divisible by both 3 and 5")
```

###2.Check whether a given number is multiple of five or not. 

```python
num = 25

if num % 5 == 0:
    print("Number is a multiple of 5")
else:
    print("Number is not a multiple of 5")
```

###3.Find the greatest among two numbers. If numbers are equal than print “numbers are equal”. 

```python
a = 10
b = 20

if a > b:
    print("a is greater")
elif b > a:
    print("b is greater")
else:
    print("numbers are equal")
```

###4.Find the greatest among three numbers assuming no two values are same.  

```python
a = 10
b = 25
c = 18

if a > b and a > c:
    print("a is greatest")
elif b > a and b > c:
    print("b is greatest")
else:
    print("c is greatest")
```

###5.Check whether the quadratic equation has real roots or imaginary roots. Display the roots. 

```python
a = 1
b = -3
c = 2

D = b*b - 4*a*c

if D > 0:
    print("Roots are Real and Different")
    root1 = (-b + D**0.5) / (2*a)
    root2 = (-b - D**0.5) / (2*a)
    print("Root 1 =", root1)
    print("Root 2 =", root2)

elif D == 0:
    print("Roots are Real and Equal")
    root = -b / (2*a)
    print("Root =", root)

else:
    print("Roots are Imaginary")
```

###6.Find whether a given year is a leap year or not. 

```python
year = 2024

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    print("Leap Year")
else:
    print("Not a Leap Year")
```

###7.Write a program which takes any date as input and display next date of the calendar e.g. 
I/P: day=20 month=9 year=2005  
O/P: day=21 month=9 year 2005 


```python
day = 20
month = 9
year = 2005


days_in_month = [31, 28, 31, 30, 31, 30, 
                 31, 31, 30, 31, 30, 31]

if (year % 400 == 0) or (year % 4 == 0 and year % 100 != 0):
    days_in_month[1] = 29

day = day + 1

if day > days_in_month[month - 1]:
    day = 1
    month = month + 1

    if month > 12:
        month = 1
        year = year + 1

print("day =", day)
print("month =", month)
print("year =", year)
```

###8.Print the grade sheet of a student for the given range of cgpa. Scan marks of five subjects and calculate the percentage. 
CGPA=percentage/10 
CGPA range: 
0 to 3.4 -> F 
3.5 to 5.0->C+ 
5.1 to 6->B 
6.1 to 7-> B+ 
7.1 to 8-> A 
8.1 to 9->A+ 
9.1 to 10-> O (Outstanding) 
 
Sample Gradesheet 
Name: Rohit Sharma 
Roll Number: R17234512 SAPID: 50005673 
Sem: 1 Course: B.Tech. CSE AI&ML 
 
Subject name: Marks 
PDS: 70 
Python: 80 
Chemistry: 90 
English: 60 
Physics: 50 
Percentage: 70% 
CGPA:7.0 
Grade:  A


```python
name = "Rohit Sharma"
roll_no = "R17234512"
sapid = 50005673
sem = 1
course = "B.Tech. CSE AI&ML"


pds = 70
python = 80
chemistry = 90
english = 60
physics = 50


total = pds + python + chemistry + english + physics
percentage = total / 5


cgpa = percentage / 10


if 0 <= cgpa <= 3.4:
    grade = "F"
elif 3.5 <= cgpa <= 5.0:
    grade = "C+"
elif 5.1 <= cgpa <= 6.0:
    grade = "B"
elif 6.1 <= cgpa <= 7.0:
    grade = "B+"
elif 7.1 <= cgpa <= 8.0:
    grade = "A"
elif 8.1 <= cgpa <= 9.0:
    grade = "A+"
else:
    grade = "O (Outstanding)"


print("----- GRADE SHEET -----")
print("Name:", name)
print("Roll Number:", roll_no, "SAPID:", sapid)
print("Sem:", sem, "Course:", course)

print("\nSubject name: Marks")
print("PDS:", pds)
print("Python:", python)
print("Chemistry:", chemistry)
print("English:", english)
print("Physics:", physics)

print("\nPercentage:", percentage, "%")
print("CGPA:", cgpa)
print("Grade:", grade)
```
