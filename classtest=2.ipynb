##Q1. Class Variable vs Instance Variable

```python
class student:
    school_name = "upes"

    def __init__(self, name, age):

        self.name = name 
        self.age = age

student1 = student("aman", 20)
student2 = student("riya", 21)

print("school1: ", student1.school_name,student1.age)
print("school2: ", student2.school_name,student2.age)

print("student 1:", student1.name, student2.age)
print("student 2:", student2.name, student2.age)
```

##Q2Instance Method and Class Method

```python
class employee:

    company ="ABC ltd"
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def show_details(self):
        print("name:", self.name, "|salary:", self.salary, "| company:", employee.company)

    @classmethod
    def change_company(cls, new_name):
        cls.company = new_name

emp1 = employee("aman", 30000)  
emp2 = employee("riya", 40000)  

print("before change:")
emp1 .show_details()
emp2 .show_details()

employee.change_company("XYZ ltd")

emp1 .show_details()
emp2 .show_details()
```

##Q4. Public and Private Data Members

```python
class Bankaccount:

    def __init__(self, name, balance):
        self.name = name

        self.__balance = balance

    def deposit(self, amount):
        self.__balance += amount

        print("deposited:", amount)

    def withdraw(self, amount):
        if amount <= self.__balance:
            self.__balance -= amount
            print("withdraw:", amount)
        else:
            print("insufficient balance")

    def show_balance(self):
        print("name:", self.name, "| balance:", self.__balance)

acc = Bankaccount("aman",  1000)


acc.deposit(500)
acc.show_balance()

acc.withdraw(300)
acc.show_balance()

acc.withdraw(1500)
```

##Q6. Magic Methods / Operator Overloading

```python
class Product:
    def __init__(self, name, price):
        self.name = name
        self.price = price

    def __str__(self):
        return f"Product: {self.name}, Price: {self.price}"

    def __add__(self, other):
        return self.price + other.price

    def __eq__(self, other):
        return self.price == other.price


# Creating objects
p1 = Product("Laptop", 50000)
p2 = Product("Phone", 30000)
p3 = Product("Tablet", 50000)

# Printing results
print(p1)
print(p1 + p2)
print(p1 == p3)
print(p1 == p2)
```
