###1.Write a Python function to find the maximum and minimum numbers from a sequence of numbers.  (Note: Do not use built-in functions.) 

```python
def find_max_min(numbers):
    maximum = numbers[0]
    minimum = numbers[0]
    
    for num in numbers:
        if num > maximum:
            maximum = num
        if num < minimum:
            minimum = num
    
    return maximum, minimum


nums = [12, 45, 7, 89, 23, 4]

max_value, min_value = find_max_min(nums)

print("Maximum:", max_value)
print("Minimum:", min_value)
```

###2.Write a Python function that takes a positive integer and returns the sum of the cube of all the positive integers smaller than the specified number.  

```python
def sum_of_cubes(n):
    total = 0
    for i in range(1, n):
        total = total + (i ** 3)
    return total

# Added value here
num = 6

result = sum_of_cubes(num)
print("Given number:", num)
print("Sum of cubes of numbers smaller than", num, "is:", result)
```

###3.Write a Python function to print 1 to n using recursion. (Note: Do not use loop) 

```python
def print_numbers(n):
    if n == 0:     
        return
    print_numbers(n - 1)   
    print(n)              


num = 5
print("Numbers from 1 to", num, "are:")
print_numbers(num)
```

###4.Write a recursive function to print Fibonacci series upto n terms. 

```python
def fibonacci(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibonacci(n-1) + fibonacci(n-2)


def print_fibonacci_terms(n, current=0):
    if current == n:
        return
    print(fibonacci(current), end=" ")
    print_fibonacci_terms(n, current + 1)


num = 7
print("Fibonacci series up to", num, "terms:")
print_fibonacci_terms(num)
```

###5.Write a lambda function to find volume of cone. 

```python
volume_cone = lambda r, h: (1/3) * 3.14159 * r**2 * h

radius = 3
height = 5

result = volume_cone(radius, height)

print("Radius:", radius)
print("Height:", height)
print("Volume of Cone:", result)
```

###6.Write a lambda function which gives tuple of max and min from a list. 
Sample input: [10, 6, 8, 90, 12, 56] 
Sample output: (90,6) 


```python
max_min = lambda lst: (max(lst), min(lst))

numbers = [10, 6, 8, 90, 12, 56]

result = max_min(numbers)

print("List:", numbers)
print("Output (Max, Min):", result)
```
